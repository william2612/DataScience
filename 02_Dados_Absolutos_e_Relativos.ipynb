## Teoria Dados Absolutos e Relativos

A coleta de amostras de uma população pode ser:

Dados absolutos: não sofreram manipulação, no máximo contagem e ordenação. <br>
Dados relativos: fácil entendimento, ajuda na comparação entre quantidades <br>
    Percentil
    Índices
    Coeficientes
    Taxas

```python
import pandas as pd
import numpy as np
```

```python
dados={'emprego':['Administrador','Programador','Executivo'],
      'São Paulo':[700, 300, 45],
      'Minas Gerais':[1200,200,9]
      }
type(dados)
```

```python
dataset=pd.DataFrame(dados)
dataset
```

```python
dataset['São Paulo'].sum()
```

```python
dataset['Minas Gerais'].sum()
```

```python
dataset['%_SP']=(dataset['São Paulo'])/dataset['São Paulo'].sum()
```

```python
dataset
```

```python
dataset['%_SP']*=100
dataset
```

```python
dataset['%_MG']=(dataset['São Paulo'])/dataset['São Paulo'].sum()*100
```

### Coeficiente e Taxa de Variação

```python
tabela={'Ano Escolar':['1°','2°','3°','4°'],'matriculas março':[70,50,47,23],'matriculas abril':[65,48,40,22]}
tabela
```

```python
dataset_tabela=pd.DataFrame(tabela)
dataset_tabela
```

A ideia é que o coeficiente de variação pelo total, a taxa de variação pode ser estudada em cima de um montante, nesse caso abaixo temos o 1° ano escolar com uma taxa de desistência de 7.14 a cada 100 pessoas ou 70 a cada 1000, ou seja é uma portagem do quando diminuiu em relação a março <br>

taxa desistência= variação março - abril/matricula março * 100 <br>
coeficiente= variação março - abril/matricula março 



```python
dataset_tabela['taxa desistencia']=(dataset_tabela['matriculas março']-dataset_tabela['matriculas abril']).abs()/dataset_tabela['matriculas março']*100
```
