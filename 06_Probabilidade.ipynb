## Probabilidade Distribuição Binomial

Ideal para elementos que são independentes. Exemplo, lançar moedas, cada lançamento não vai afetar a chance dos demais.

Tirar 5 moedas:

X=5 número de sucesso
p=  probabilidade de sucesso (0,5)
n=10 (quantidade de tentativas ou trials)






p= PEÇA NÃO TER DEFEITOS É DE 99% => ¬p=1%
n=30 peças 
k= 28 em boas condições 

temos que f(k)=P(X=k) = a probabilidade para 28 boas condições

P(X >=28)= P(28) + P(29) + P(30)

=0.99
P(28)=30!/28!(30-28)! x p^(28)x (1-p)^(30-28)=0,0328
P(29)=30!/29!(30-29)! x p^(29)x (1-p)^(30-29)=0,224
P(30)=30!/30!(30-30)! x p^(30)x (1-p)^(30-30)=0,7397


P(X>=28)=0,997

podemos indagar o seguinte: qual a chance de haver menos do que 28 peças com defeito? 1-P(X>=28) é o complementar e responderia essa pergunta
P(X<28)

MEDIA= n x p= 30x 99%=29,7 <br>
VAR= n x p (1-p)=29,7x(0.01)=0,297

```python
n = 10
x = 5
p = 0.5
```

```python
import math
(math.factorial(n)) / (math.factorial(x) * math.factorial(n - x)) * math.pow(p, x) * math.pow(1 - p, n - x)
```

```python
scp.stats.binom.pmf(x, n, p) #probability mass funcion
```

## Distribuição de Poisson

Utilizada para números discretos, quantidade finita de ocorrência com base numa outra variável contínua.

Não leva em conta o número de experimentos.

Os eventos precisam ser independentes:



Na distribuição de poisson a média é igual a variância:

E(X)=Var(X)=Lambda

Exemplo 1:
Suponha que o número de clientes que chegam a um banco siga uma distribuição Poisson.Verifica-se que, em
média, chegam 12 clientes por minuto. Calcule: a) probabilidade de chegada de 10 clientes no próximo minuto;
b) probabilidade de chegada de 40 clientes nos próximos 5 minutos; c) média e variância de X. 

Solução:

A)= P(X=k)=>P(X=10)= (e^-12 x 12^10 )/10! =10.483725588365921 %
B)=> 40/5= 8 clientes por minutos,  implica em nossa observção k=> P(X=8)=6.552328492728705 %
C)=> 12

Exemplo 2:

Determinada peça é produzida em uma linha de produção.A probabilidade de que essa peça seja defeituosa é
de 0,01. Se forem produzidas 300 peças, qual a probabilidade de que nenhuma delas seja defeituosa?


Solução: podemos considerar que n é muito grande em relação a p muito pequeno, logo podemos aproximar a distribuição binomial para

media=0,01 x 300= 3 peças defeitusas 

logo 

x= 0 deifeitos/peça
media= 3 defeitos a cada 300 peças

P(X=0)=4.978706836786395




```python
x = 0
media = 3
scp.stats.poisson.pmf(x, media) * 100
```
