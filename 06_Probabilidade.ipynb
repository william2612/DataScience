## Probabilidade Distribuio Binomial

Ideal para elementos que so independentes. Exemplo, lanar moedas, cada lanamento no vai afetar a chance dos demais.

Tirar 5 moedas:

X=5 nmero de sucesso
p=  probabilidade de sucesso (0,5)
n=10 (quantidade de tentativas ou trials)






p= PEA NO TER DEFEITOS  DE 99% => p=1%
n=30 peas 
k= 28 em boas condies 

temos que f(k)=P(X=k) = a probabilidade para 28 boas condies

P(X >=28)= P(28) + P(29) + P(30)

=0.99
P(28)=30!/28!(30-28)! x p^(28)x (1-p)^(30-28)=0,0328
P(29)=30!/29!(30-29)! x p^(29)x (1-p)^(30-29)=0,224
P(30)=30!/30!(30-30)! x p^(30)x (1-p)^(30-30)=0,7397


P(X>=28)=0,997

podemos indagar o seguinte: qual a chance de haver menos do que 28 peas com defeito? 1-P(X>=28)  o complementar e responderia essa pergunta
P(X<28)

MEDIA= n x p= 30x 99%=29,7 <br>
VAR= n x p (1-p)=29,7x(0.01)=0,297

```python
n = 10
x = 5
p = 0.5
```

```python
import math
(math.factorial(n)) / (math.factorial(x) * math.factorial(n - x)) * math.pow(p, x) * math.pow(1 - p, n - x)
```

```python
scp.stats.binom.pmf(x, n, p) #probability mass funcion
```

Utilizada para nmeros discretos, quantidade finita de ocorrncia com base numa outra varivel contnua.

No leva em conta o nmero de experimentos.

Os eventos precisam ser independentes:



Na distribuio de poisson a mdia  igual a varincia:

E(X)=Var(X)=Lambda

b) probabilidade de chegada de 40 clientes nos prximos 5 minutos; c) mdia e varincia de X. 

Soluo:

A)= P(X=k)=>P(X=10)= (e^-12 x 12^10 )/10! =10.483725588365921 %

Exemplo 2:
